## 1. Булеві значення та квазіістинність

### Тип bool

`bool` у Python — це **підклас** цілого числа `int`:

| Значення | Числовий еквівалент |
|----------|---------------------|
| `True`   | `1`                 |
| `False`  | `0`                 |

Тому арифметика з `bool` цілком законна:

```python
True + True   # → 2
True + False  # → 1
```

---

### Truthy та Falsy об'єкти

Кожен об'єкт у Python неявно оцінюється як `True` або `False` в умовних виразах.

**Завжди False (Falsy):**

| Значення | Тип      |
|----------|----------|
| `0`      | int      |
| `0.0`    | float    |
| `""`     | str      |
| `[]`     | list     |
| `{}`     | dict     |
| `None`   | NoneType |

**Все інше** — `True` (Truthy): `"False"`, `-1`, `[0]`, `" "` тощо.

---

## 2. Логічні оператори та пріоритет

| Оператор | Опис                                       |
|----------|-----------------------------------------|
| `not`    | заперечення: `not True` → `False`        |
| `and`    | обидва True → True                       |
| `or`     | хоча б один True → True                  |

### Пріоритет (від вищого до нижчого)

```
not  →  and  →  or
```

Тому вираз `not x and y or z` читається як `((not x) and y) or z`.

---

### Short-circuiting (укорочена оцінка)

Python зупиняє обчислення, як тільки результат стає очевидним:

```python
True or f()    # f() НЕ викликається
False and f()  # f() НЕ викликається
```

Практичне застосування — безпечне ділення:

```python
if y != 0 and x / y > 1:   # ділення не відбудеться якщо y == 0
    ...
```

---

## 3. None — відсутність значення

`None` — спеціальний об'єкт типу `NoneType`:

- Існує **лише один** `None` в Python (singleton)
- Означає «значення відсутнє» (аналог `null` в інших мовах)
- Функція без `return` **завжди повертає `None`**

```python
def greet():
    print("Hi")

result = greet()   # result == None
```

**Правильна перевірка:**

```python
if value is None:    # ✅ рекомендовано
    ...

if value == None:    # ⚠️ працює, але не рекомендовано
    ...
```

## 4. Умовні конструкції: if / elif / else

```python
if умова1:
    # виконується якщо умова1 True
elif умова2:
    # виконується якщо умова2 True (і умова1 False)
else:
    # виконується якщо жодна умова не True
```

**Ключове правило:** Python перевіряє умови **зверху вниз**.  
Як тільки знаходить першу `True` — виконує блок і виходить з конструкції.

```python
score = 85
if score >= 60:    # ← True! Виконується цей блок.
    print("Склав")
elif score >= 80:  # ← ніколи не перевіряється
    print("B")
```

## 5. Цикл while

```python
while умова:
    # тіло циклу
    # ← обов'язково змінювати умову!
```

Підходить коли кількість ітерацій **наперед невідома**.

**Типова помилка — нескінченний цикл:**

```python
x = 1
while x <= 5:
    print(x)   # x ніколи не змінюється → зависання
```

**Правильно:**

```python
x = 1
while x <= 5:
    print(x)
    x += 1     # наближаємо умову до False
```

## 6. Управління циклом: break та continue

| Оператор   | Що робить                         |
|------------|-----------------------------------|
| `break`    | повністю зупиняє цикл             |
| `continue` | пропускає поточну ітерацію        |

```python
for i in range(1, 6):
    if i == 3:
        continue  # пропустити 3
    if i == 5:
        break     # зупинитись перед 5
    print(i)
```

## 7. Потік виконання: `__name__ == "__main__"`
Уявіть, що ваш файл із кодом — це людина, яка може виступати у двох різних ролях:
1. **"Господар"** — коли ви запускаєте саме цей файл напряму (як самостійну програму).
2. **"Гість"** — коли інший файл підключає його до себе за допомогою команди `import`, щоб просто скористатися його функціями.

Спеціальна вбудована змінна `__name__` — це своєрідний "бейдж", який інтерпретатор Python автоматично чіпляє на ваш файл, щоб файл розумів, у якій ролі він зараз перебуває.

### Як змінюється цей "бейдж":
* Коли ви **запускаєте файл напряму** (наприклад, через термінал `python my_program.py` або натиснувши кнопку Run), Python присвоює змінній `__name__` значення `"__main__"`.
* Коли ви **імпортуєте файл** в інший скрипт (наприклад, `import my_program`), Python присвоює цій змінній значення, що відповідає реальній назві файлу (наприклад, `"my_program"`).

### Що робить рядок `if __name__ == "__main__":`?
Це звичайна умова, яка буквально перевіряє: *"Цей файл зараз запустили як головну програму?"*.
* Якщо **ТАК** (умова виконується), то викликається функція `main()` або будь-який інший код, схований під цим `if`.
* Якщо **НІ** (файл просто імпортували), блок коду під `if` повністю ігнорується.

### Чому це вважається найкращою практикою?
У Python під час імпорту файлу інтерпретатор прочитує його і **одразу виконує всі інструкції**, які знаходяться на верхньому рівні.

Якби ви не сховали головний запуск програми під `if __name__ == "__main__":`, то при спробі просто імпортувати ваш файл з іншої програми, ваш код почав би раптово виконуватися: щось би друкувалося на екран, запускалася б логіка гри або тести. Інша програма просто хотіла б "позичити" у вас пару функцій, а натомість отримала б купу непотрібних дій.
---
Використовуючи цю конструкцію, ви робите свій код універсальним. Він працює **у двох режимах**: і як самостійний скрипт, який виконує корисну роботу при прямому запуску, і як тиха "бібліотека" інструментів (модуль), яка нічого зайвого не робить, коли її просто імпортують.
---

```python
if __name__ == "__main__":
    main()
```

Python присвоює `__name__`:

| Сценарій             | `__name__`     | Блок виконується? |
|----------------------|----------------|-------------------|
| Запуск напряму       | `"__main__"`   | ✅ так            |
| Імпорт як модуль     | назва файлу    | ❌ ні             |

Це найкраща практика для структурування Python-програм.

```python
while x <= 5:
```

### Що виведе Python?

```python
for num in range(1, 6):

        continue
    if num == 5:
        break
    print(num)
```

````python
#@title 🔮 Ваш прогноз
#
# for num in range(1, 6):
#     if num == 3: continue
#     if num == 5: break
#     print(num)
#
# Введіть числа через пробіл:

if not require_student():
    raise SystemExit

prediction_loop = "" #@param {type:"string"}

if prediction_loop.strip() == "":
    print("⚠️ Введіть прогноз перед виконанням наступної клітинки")
    raise SystemExit

print("✅ Прогноз збережено:", prediction_loop)
````

````python
#@title 🐍 Результат та пояснення запусти Run

from IPython.display import Markdown, display

if not require_student():
    raise SystemExit

try:
    print(f"{STUDENT_NAME}, твій прогноз: {prediction_loop}")
except NameError:
    raise ValueError("⚠️ Спочатку зробіть прогноз у клітинці 🔮")

# ── EXECUTION ──────────────────────────────────────────

print("\n🐍 Python виводить (покроково):")
for num in range(1, 6):
    if num == 3:
        print(f"  num={num} → continue (пропускаємо)")
        continue
    if num == 5:
        print(f"  num={num} → break (зупиняємось)")
        break
    print(f"  num={num} → print({num})")

print("\n📌 Фактичний вивід:")
for num in range(1, 6):
    if num == 3:
        continue
    if num == 5:
        break
    print(num)

# ── AUTO CHECK ─────────────────────────────────────────

correct_answer = "1 2 4"
if prediction_loop.strip() == correct_answer:
    print("\n✅ Правильний прогноз!")
else:
    print("\n❌ Неправильно. Очікувалось:", correct_answer)

# ── ПОЯСНЕННЯ ──────────────────────────────────────────

display(Markdown("""
---

## ✅ Що сталося?

| Ітерація | num | Дія          | Виводить? |
|----------|-----|--------------|----------|
| 1        | 1   | звичайна     | ✅ `1`   |
| 2        | 2   | звичайна     | ✅ `2`   |
| 3        | 3   | `continue`   | ❌ пропуск |
| 4        | 4   | звичайна     | ✅ `4`   |
| 5        | 5   | `break`      | ❌ вихід |

**Результат:** `1 2 4`

---

### break vs continue

```python
# break — повністю зупиняє цикл
for i in range(10):
    if i == 3:
        break   # вихід: виведе 0 1 2
    print(i)

# continue — пропускає лише цю ітерацію
for i in range(5):
    if i == 3:
        continue  # пропуск: виведе 0 1 2 4
    print(i)
```

---

## 🔑 Головне правило

> `break` = вийти з циклу назавжди.  
> `continue` = перейти до наступної ітерації.
"""))

# ── SUBMIT ─────────────────────────────────────────────

try:
    submit_task(STUDENT_NAME, "Lesson2", "Task_break_continue")
except:
    print("\nℹ️ submit_task() не підключений")
````

# 7. Потік виконання: `__name__ == "__main__"`

## Навіщо це потрібно?

Ця конструкція дозволяє Python розрізняти два сценарії:

| Сценарій              | `__name__`     | Блок виконується? |
|-----------------------|----------------|-------------------|
| Запуск напряму        | `"__main__"`   | ✅ так            |
| Імпорт з іншого файлу | назва файлу    | ❌ ні             |

```python
# my_module.py

def greet(name):
    return f"Hello, {name}!"

if __name__ == "__main__":
    # виконується тільки при прямому запуску
    print(greet("World"))
```

Якщо інший файл зробить `import my_module` — функція `greet` буде доступна, але виклик у блоці **не виконається**.

````python
# Демонстрація: що таке __name__ у поточному середовищі

print("Значення __name__:", __name__)
print()

if __name__ == "__main__":
    print("✅ Файл запущено напряму")
else:
    print("📦 Файл імпортовано як модуль")
````
