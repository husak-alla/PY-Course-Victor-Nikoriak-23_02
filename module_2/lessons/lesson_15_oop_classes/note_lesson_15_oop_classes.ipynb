## 🔬 Частина 1: Все є об'єктом

У мовах як Java або C є **примітивні типи**: `int`, `float`, `bool` — вони не є об'єктами, не мають методів.  
Python влаштований інакше.

### ❓ Запитання / Передбачення

Як ви думаєте — чим є звичайне число `5` для Python?  
Просто коміркою пам'яті? Або щось більше?

Запустіть код нижче і подивіться на результат.


````python
# Чи є звичайні типи даних об'єктами?
print(isinstance(5, object))           # число
print(isinstance('hello', object))     # рядок
print(isinstance([1, 2, 3], object))   # список
print(isinstance(True, object))        # булевий
````

**✅ Результат:** `True`, `True`, `True`, `True`

**Пояснення:**  
У Python **немає примітивних типів** у класичному розумінні.  
Число `5` — це об'єкт. Він має методи, атрибути, і живе в ієрархії класів.  
Усе, з чим ви працюєте в Python, є нащадком єдиного базового класу `object`.

> 🔑 **Ментальна модель:**  
> Змінна — це не "коробка з числом".  
> Змінна — це **ярлик** (посилання), що вказує на **об'єкт** у пам'яті.


````python
# Доказ: дві змінні — один об'єкт
a = [1, 2, 3]
b = a           # b — це НЕ копія, це ще один ярлик на той самий список

b.append(4)

print('a =', a)        # [1, 2, 3, 4] — змінилось!
print('b =', b)        # [1, 2, 3, 4]
print(id(a) == id(b))  # True — один і той самий об'єкт у пам'яті
````

> 🧠 **Чому `a` змінився?**  
> Тому що `a` і `b` — це два ярлики, наклеєних на **один і той самий список**.  
> Ми нічого не копіювали. Якщо хочете копію — використовуйте `b = a.copy()`.

## 🧬 Частина 2: Звідки беруться класи?

Ми з'ясували, що `5` — це об'єкт класу `int`.  
Але звідки береться сам клас `int`?

### ❓ Запитання / Передбачення

Якщо `5` — об'єкт, що має клас `int`,  
то `int` — теж якийсь об'єкт. Але якого класу?

Запустіть код і спробуйте усвідомити результат.


````python
print(type(5))      # клас числа 5
print(type(int))    # клас самого класу int
print(type(type))   # клас класу type
````

**✅ Результат:**
```
<class 'int'>
<class 'type'>
<class 'type'>
```

**Пояснення (крок за кроком):**

| Вираз | Результат | Що це означає |
|-------|-----------|---------------|
| `type(5)` | `int` | Число `5` створене класом `int` |
| `type(int)` | `type` | Клас `int` створений класом `type` |
| `type(type)` | `type` | Клас `type` створив **сам себе** |

**`type`** — це **метаклас**: клас, який створює класи.  
Python сам себе описує. Це рекурсивна, але повністю узгоджена система.

---

**Дерево ієрархії (спрощено):**
```
         object          ← прабатько всього
        /  |  \  \
      int str list  ...
       |             
       5  'hello'  []
```

Усі класи (вбудовані та ваші власні) є нащадками `object`:


````python
print(issubclass(int, object))    # True
print(issubclass(str, object))    # True
print(issubclass(list, object))   # True
````

## 🏗️ Частина 3: Клас vs Екземпляр

**Клас (Class)** — це **креслення** (blueprint).  
Він описує, які дані та поведінку матимуть об'єкти, але сам нічого не "робить".

**Екземпляр (Instance)** — це **конкретний об'єкт**, створений за кресленням і розміщений у пам'яті.

```
Клас Dog        →    Екземпляри
────────────        ─────────────────────
Ім'я: ???           dog1.name = 'Рекс'
Порода: ???         dog2.name = 'Лайка'
[ говорити() ]      dog3.name = 'Шарик'
```

Аналогія: клас `Dog` — як **форма для печива** 🍪.  
Кожне печиво (екземпляр) має однакову форму, але може бути різного розміру і смаку.


````python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def speak(self):
        print(f'{self.name} каже: Гав!')


# Створюємо два різних екземпляри одного класу
dog1 = Dog('Рекс', 'Вівчарка')
dog2 = Dog('Лайка', 'Лайка')

dog1.speak()
dog2.speak()

print(type(dog1))              # <class '__main__.Dog'>
print(isinstance(dog1, Dog))   # True
print(isinstance(dog1, object)) # True — теж нащадок object!
````

## ⚙️ Частина 4: `__init__` та `self`

### Що таке `__init__`?

`__init__` — це **ініціалізатор** об'єкта.  
Python викликає його автоматично, щойно ви пишете `Dog('Рекс', 'Вівчарка')`.  
Об'єкт вже створено в пам'яті — `__init__` лише **наповнює його даними**.

### Що таке `self`?

`self` — це **посилання на конкретний екземпляр**, з яким зараз працює метод.

Коли ви пишете:
```python
dog1.speak()
```
Python насправді перетворює це на:
```python
Dog.speak(dog1)   # ← dog1 передається як перший аргумент (self)
```

Без `self` метод не знав би, **до якого саме екземпляра** він звертається.

### ❓ Запитання

Що виведе цей код? Чому?


````python
class Robot:
    def __init__(self, name, version):
        self.name = name
        self.version = version

    def introduce(self):
        print(f'Я {self.name}, версія {self.version}.')

    def __str__(self):
        # __str__ викликається коли ви робите print(об'єкт)
        return f'<Robot: {self.name} v{self.version}>'


r1 = Robot('R2D2', 2)
r2 = Robot('C3PO', 3)

r1.introduce()
r2.introduce()

print(r1)   # викликає __str__
print(r2)   # викликає __str__
````

**Дандер-методи (`__dunder__`)** — це спеціальні методи Python з подвійним підкресленням.  
Вони дають вашим об'єктам можливість взаємодіяти з вбудованими операціями Python.

| Метод | Коли викликається |
|-------|------------------|
| `__init__` | При створенні об'єкта `obj = MyClass()` |
| `__str__` | При `print(obj)` або `str(obj)` |
| `__len__` | При `len(obj)` |
| `__repr__` | При виводі в консолі або `repr(obj)` |

## 📦 Частина 5: Атрибути класу vs атрибути екземпляра

У Python є **два рівні** зберігання атрибутів:

| Тип | Де зберігається | Спільний чи особистий? |
|-----|-----------------|------------------------|
| **Атрибут класу** | В самому класі | **Спільний** для всіх екземплярів |
| **Атрибут екземпляра** | В `self` (об'єкті) | **Особистий** кожного екземпляра |

### ❓ Передбачення

Що виведе останній рядок?


````python
class Dog:
    species = 'Canis lupus familiaris'  # атрибут КЛАСУ (спільний)

    def __init__(self, name):
        self.name = name                 # атрибут ЕКЗЕМПЛЯРА (особистий)


dog1 = Dog('Рекс')
dog2 = Dog('Лайка')

# Змінюємо атрибут через КЛАС — змінюється для всіх
Dog.species = 'Вовк'

# Змінюємо через ЕКЗЕМПЛЯР — створюється особистий атрибут лише у dog1
dog1.species = 'Прибулець'

print(dog1.species)   # ?
print(dog2.species)   # ?
print(Dog.species)    # ?
````

**✅ Результат:**
```
Прибулець
Вовк
Вовк
```

**Пояснення — порядок пошуку атрибута:**

```
dog1.species  →  Шукаємо в dog1.__dict__  →  Знайдено: 'Прибулець' ✅
dog2.species  →  Шукаємо в dog2.__dict__  →  Не знайдено
              →  Шукаємо в Dog.__dict__   →  Знайдено: 'Вовк' ✅
```

> ⚠️ **Важливо!**  
> Коли ви пишете `dog1.species = 'Прибулець'`, ви **не змінюєте** атрибут класу.  
> Ви створюєте **новий особистий атрибут** у словнику `dog1.__dict__`.  
> Клас `Dog` і `dog2` залишаються незачепленими.

Перевірити словники об'єктів можна так:


````python
class Dog:
    species = 'Canis'
    def __init__(self, name):
        self.name = name

dog1 = Dog('Рекс')
dog1.species = 'Прибулець'

print('dog1.__dict__:', dog1.__dict__)   # особисті атрибути
print('Dog.__dict__ keys:', list(Dog.__dict__.keys()))  # атрибути класу
````

## 🔍 Частина 6: `isinstance` vs `type` vs `issubclass`

Python надає три інструменти для перевірки типів. Вони схожі, але використовуються в різних ситуаціях.


````python
class Animal:
    pass

class Bird(Animal):   # Bird наслідує Animal
    pass

class Fish(Animal):   # Fish наслідує Animal
    pass


duck = Bird()
salmon = Fish()

# --- type() ---
# Повертає ТОЧНИЙ клас об'єкта (без урахування наслідування)
print(type(duck) is Bird)     # True
print(type(duck) is Animal)   # False! (duck — це Bird, не Animal)

# --- isinstance() ---
# Перевіряє: чи є об'єкт екземпляром цього класу АБО будь-якого його батька
print(isinstance(duck, Bird))    # True
print(isinstance(duck, Animal))  # True  ← знає про наслідування!
print(isinstance(duck, Fish))    # False

# --- issubclass() ---
# Перевіряє зв'язок між двома КЛАСАМИ (не екземплярами)
print(issubclass(Bird, Animal))  # True
print(issubclass(Fish, Animal))  # True
print(issubclass(Bird, Fish))    # False
````

**Коли що використовувати:**

| Функція | Аргументи | Використовуйте коли... |
|---------|-----------|------------------------|
| `type(x) is Cls` | об'єкт | Потрібен **суворо** цей клас (без нащадків) |
| `isinstance(x, Cls)` | об'єкт, клас | Об'єкт **є або наслідує** цей клас (95% випадків) |
| `issubclass(A, B)` | клас, клас | Перевіряєте зв'язок між двома **класами** |

> ✅ **Правило:**  
> У реальному коді майже завжди використовуйте `isinstance`.  
> Воно поважає ієрархію наслідування і є основою **поліморфізму**.

## 🔭 Частина 7: LEGB та класи

Python шукає змінні за правилом **LEGB**:
```
L — Local      (всередині функції/методу)
E — Enclosing  (в охоплюючій функції, для вкладених функцій)
G — Global     (на рівні модуля)
B — Built-in   (вбудовані: print, len, range...)
```

⚠️ **Важливо:** простір імен **класу** не входить до LEGB!  
Методи класу **не бачать** атрибути класу напряму — лише через `self` або `ClassName`.

### ❓ Передбачення

Що виведуть три `print`?


````python
X = 'глобальна'   # Global

class MyClass:
    X = 'класова'   # Простір імен класу (НЕ є Enclosing для методів!)

    def show(self):
        X = 'локальна'   # Local

        print(X)         # ? (шукає Local → знаходить 'локальна')
        print(self.X)    # ? (звертаємось до класу через self)

    def show_global(self):
        # Тут немає локальної X
        print(X)         # ? (шукає Local → нема, Enclosing → нема,
                         #    Global → знаходить 'глобальна'!)
                         #    Клас ПРОПУСКАЄТЬСЯ!


obj = MyClass()
obj.show()
obj.show_global()
````

**✅ Результат:**
```
локальна
класова
глобальна
```

> 💡 **Висновок:**  
> Щоб звернутися до атрибута класу з методу — завжди використовуйте `self.X` або `MyClass.X`.  
> Без кваліфікатора Python не буде шукати в просторі імен класу.
