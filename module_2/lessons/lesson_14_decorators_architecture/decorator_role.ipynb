```python
import  functools


def admin (*role):
    """ Приймає дозволені ролі """

    def decorator(func):

        @functools.wraps(func)

        def wrapper(*args, **kwargs):
            if current_user["role"] not in role:
                print("Доступ заборонено")
                return
            return func(*args, **kwargs)
        return wrapper
    return decorator

```

```python
@admin("guest", "user", "admin", "moderator")
def view_post(post_id):
    print(f"{current_user["name"]} дивиться #{post_id}")
```

```python
@admin("admin")
def delete_post(post_id):
    print(f"{current_user["name"]} удаляє #{post_id}")
```

```python
@admin("user", "admin", "moderator")
def create_post(title):
    print(f"{current_user["name"]} створює пост {title}")
```
